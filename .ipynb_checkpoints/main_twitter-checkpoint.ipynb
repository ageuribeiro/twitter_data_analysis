```python
#Função para remover Stopwords da nossa base:
def RemoveStopWords(instancia):
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in instancia.split() if not i in stopwords]
    return (" ".join(palavras))
```

<h1>Stemming</h1>

<p>É a técnica de reduzir uma palavra para o seu radical, mesmo que o radical em si não seja uma palavra válida no dioma. Removendo os sufixos e prefixos de uma palavra.

<ul>
    <li>Exemplo: o radical da palavra ilegal é ileg.</li>
</ul>
    </p>

```python
#Aplicando o stemming em nossa base
def Steammin(instancia):
    stemmer = nltk.stem.RSLPStemmer()
    palavras = []
    for w in instancia.split():
        palavras.append(stemmer.stem(w))
    return(" ".join(palavras))
```

<h2>Remove Caracteres Indesejados Como Links,etc.</h2>

```python
import re
```

```python
def limpeza_dados(instancia):
    #remove links, pontos, vírgulas, ponto e vírgulas dos tweets
    instancia = re.sub(r"http\s+","", instancia).lower().replace(',','').replace(';','').replace('-','').replace(':','').replace(')','')
    return(instancia)
```

<h1>Lemmatization</h1>

<p> Reduz as palavras flexionadas adequadamente, garantindo que a palavra raiz pertença ao idioma.</p>
<p> Determinando assim a palavra que representa seu lema.</p>

<h3> <b>Por exemplo execuções e xecutar, são forma da palavra execução, portanto, execução é o lema dessas palavras</b></h3>

<p><b><code>Observação: </code></b> A função será mostrada a seguir, para fins de conhecimento, pois esse processo só está disponível para o idioa inglês, que não é o nosso caso.</p>

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

def Lemmatization(instancia):
    palavras = []
    for w in instancia.split():
        palavras.append(wordnet_lemmatizer.lemmatizer(w))
    return(" ",join(palavras))
```
